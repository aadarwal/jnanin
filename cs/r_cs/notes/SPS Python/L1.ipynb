## 1. Variables

Python variables are just like variables in mathematics. They simply store information. For instance, the variable $x$ in the equation

\begin{equation}
x + 3 = 5
\end{equation}

stores information about the number ($x = 2$) that satisfies that equation. Similary in Python we can assign values to variables and Python stores that information for later use. For instance,

```python
#define the variable x_1 to be equal to 2 and x_2 to be 3
x_1 = 2 
x_2 = 3
#define the variable y to be equal to -1000
y = -1000
```

All the above did is stored the numbers 2 and 3 into the respective variables. We can make sure that they have been stored correctly by just printing the values of the variables.

```python
print(x_1)
print(y)
print(x_2)
```

These stored variables we can also do computations like ...

```python
addition = x_1 + x_2

print(addition)
```

What we did here is that we defined a new variable called "addition" and assigned it the value of x_1 + x_2. If we print out the value of the "addition" variable, we see that it is 5. This makes sense because 2 + 3 is indeed 5. 

Now if one just wanted to compute the value of $2+3$, we don't need to go through all this variable stuff and could have just done...

```python
print(2 + 3)
```

Also, just to make sure things are clear, both the below ways work ...

```python
print(x_1 + x_2)
print(addition)
```

However, for more complicated problems than this, using variables greatly simplifies stuff.

Another important thing about variables is that their values can be reassigned. That is,


```python
var1 = 1
var1 = 4
print(var1)
```

Also, reassignment can also be done like this...

```python
var2 = 3
print(var2)
var2 = var2 + 1
print(var2)
var2 = var2 + 5
print(var2)
```

Notice here that we are reassigning the variable based off of the previous value of the variable. See what happens if you repeatedly run the above cell chunk.

### Floats

"Float" is short for floating point number (decimal number). Therefore, any decimal number is a "float" or a "floating point number"

```python
x = 3.14
print(type(x))
```

```python
print(type("1"))
```

```python
#It is important to keep in mind that 2 is a different data type than 2.0.
x = 2
y = 2.0

print(type(x))
print(type(y))
```

Many times, though, floats and ints play well with each other. There is generally no problem when trying to apply operations between ints and floats. This is not true of other data types though.

Another very important kind of floats are numbers in scientific notation. 

```python
ton = 1e3 #this is 1 x 10^3  = 1000
sun_mass = 1.989e31 #mass of sun in kilogram
proton_mass = 1.672e-27 #mass of proton in kilograms

print(ton)
print(sun_mass)
print(proton_mass)

print(type(ton))
print(type(sun_mass))
print(type(proton_mass))
```

### Lists

As the name suggests lists are just "lists of things". These things could be anything, they could be variables, booleans, string, integers etc.  We can even store lists inside lists!


```python
list1 = [1,2,3,"batman","123", [1,3,4],True]

print(type(list1))
```

Now, let us consider a more useful list.

```python
list_2 = [1,4,9,16,25]

#this is the list of the first 5 squares
print(type(list_2))
```

An important feature of lists is INDEXING!

By indexing, I mean that each element in the list is assigned an index number of position. 

The first element of the list has an index number of 0. The second element of the list has an index number of 1. The third element of the list has an index number of 2. And so on...

```python
#the first element has an index number of 0

print(list_2[0])
print(list_2[1])
print(list_2[2])
print(list_2[3])
print(list_2[4])
```

```python
print(list_2[5])
```

The above error makes sense, because has the list ends at 25 and has no element with index number 5. 


Furthermore, just like how we can change the value of a variable, we can change the elements inside of a list. Consider the below example ...

```python
list_1 = [1,2,3,4,5]
```

```python
#We will change the first element just like we change a variable...
list_1[0] = 0

print(list_1)
```

We see that we changed the first element of the list from the 1 to 0!

We will talk more about lists soon. But for now, the most important thing to remember with regards to indexing: **the list index starts at 0!**

### Dictionaries

Dictionaries are just like lists, but instead of indexing them by numbers 0,1,2, ... like lists, we index them by "keys" that we provide. What I mean by that is 

```python
#a dictionary of 4 different people
ages = {"Alice" : 10, " Bob": 43, "Chris" : 13, "Dumbledore": 150}

ages["Alice"]
```

```python
#a dictionary of the information of 3 particles inside the atom.

particles = {"proton": [1,"Rutherford","p"], 
             "electron": [-1,"Thomson","e"], 
             "neutron": [0,"Chadwick","n"]}

particles["electron"]
```

You can also just print entire dictionaries.

```python
print(particles)
```

Importantly, you can also print the keys of a dictionary with dictionary_name.keys(). For example,

```python
particles.keys()
```

This is especially useful when the dictionary is extremely long and printing it is not feasible! Look at how ugly even a short dictionary prints as above.

### Tuples

Tuples are not used directly that often so I will not go in too much depth regarding what it is. The below example shows the difference between tuples and lists. 


```python
#notice the difference in brackets
list_example = [1,2,3,4,5]
tuple_example = (1,2,3,4,5)

print(list_example)
print(tuple_example)
```

We saw earlier that lists can be modified. That is, we can change the value of elements inside the list.

The same ***cannot*** be done for tuples.

Example shown below

```python
list_example[2] = 100
print(list_example)

#successfully modified the element inside list
```

```python
tuple_example[2] = 100

#cannot modify elements inside tuples
```

Do not worry about tuples too much! We won't be using them explicitly in this workshop and their knowledge is not crucial at this stage.

## 1. Functions

### Basics

Functions in Python work like how they do in math: it takes in input(s), acts on them, then returns some output(s). To translate this into code, we have to follow a specific syntax structure.
```
def func(x):
    # some step(s) that act on x
    return result
```

Notice that we use the reserved word "def" to tell Python that we are defining a function. We then must indent everything following that we wish to include as a part of the function. The reserved word "return" is usually used to end the function, and it specifies what the function will output.

*Note: We use the term "reserved word" to describe words that python has "reserved" for specific purposes (such as starting and ending a function in this example). It is important that we don't name variables after these reserved words. Some examples of reserved words are True, False, def, return*

For a more concrete example, lets consider a very simple quadratic function.

\begin{equation}
f (x) = x^2 + 5
\end{equation}

In code:

````python
def func(x):
    f = x**2 + 10 
    return(f)
````

To call this function, we simply use its name and enter the desired value of x in parentheses.

````python
func(3)
````

As you can see, this is very similar to how we notate functions in math! 

We can combine this with some of the basics from the last session, like print statements or variable assignment.

````python
# print statements
print(func(3))
print(func(5))

# variable assignment
y = func(-2)
print(y)
````

## 2. Logical Operators

We'll switch gears and return to our good-old friend, Booleans (True or False)!

A type of operation we didn't talk about is logical operators. They are like mathematical operations, but will return a Boolean instead of another number. Let's first examine inequalities.

As expected, ">" is greater than and "<" is less than. 

````python
print(10 > 1)
print(1 > 10)
print(3 < 5)
print(5 < 3)
````

If we want to test whether two values are equivalent, we use "==".

### Combining Inequalities

We can use some special reserved words to combine these inequalities

#### and

If a sequence of statements are combined with the word "and", Python will only return True if ALL the statements are True.

````python
x = 1
print(x < 5 and x < 7)
print(x < 5 and x < 0)
````

#### or

If a sequence of statements are combined with the word "or", Python will return True so long as just one of them is True.

````python
x = 2 
print(x<1 or x<3)
````

#### not

Lastly, not will reverse the result (if the result would be True, not will make it False and vice versa).

````python
x = 1
print(not(x < 0))
print(not(x > 0))
````

## 3. Conditionals

Let's use these logical operators to build more sophisticted code! We will now talk about the if-else control structure. Here is the general idea:

```
if condition_1 is True:
    # do something
else:
    # do something else
```

This is best understood through an example

````python
x = 2

if x < 3:
    print("x is dumb")
else:
    print("x is smart")
    
````

Notice, though, that we don't cover the case of if x is 0. What if we want to have multiple specific if statements? We can use the reserved word "elif" to have multiple if checks. Then, everything that isn't covered by the "if" or "elif" statements will be covered by the else.

````python
x = 1

if x > 0:
    print('x is positive')
elif x == 0:
    print('x is 0')
else:
    print('x is negative')
````

This combines really well with functions that we talked about previously.

````python
def sign_checker(x):
    if x > 0:
        print('x is positive')
    elif x == 0:
        print('x is 0')
    else:
        print('x is negative')
````

Let's test our function!

````python
sign_checker(1)
sign_checker(-1)
sign_checker(0)
````

There's one case we didn't cover for: what if x is not a number! We can add an elif statement to cover this. Recall that the special type() function returns the type of the input.

````python
def sign_checker(x):
    if x > 0:
        print('x is positive')
    elif x == 0:
        print('x is 0')
    elif type(x) != float and type(x) != int:
        print('x must be a number!')
    else:
        print('x is negative')
````

````python
sign_checker('Oops!')
````

What happened here? Remember that Python works from top to bottom. So, before we even got to the elif statement that checks the type of x, it already tries to perform x > 0. However, we cannot perform such an operation on a string! To easily fix this, we can just rearrange the function such that the type check happens first.

````python
def sign_checker(x):
    if type(x) != float and type(x) != int:
        print('x must be a number!')
    elif x > 0:
        print('x is positive')
    elif x == 0:
        print('x is 0')
    else:
        print('x is negative')
````

````python
sign_checker('Oops!')
sign_checker(5)
````

Great! Now are function is working as expected, and it can robustly handle when the input isn't a number. 

### Optional: assert

In the example above, we could have also used the keyword "assert". With assert, we can essentially force Python to have an error if some condition isn't met.

````python
def sign_checker(x):
    assert type(x) == int or type(x) == float, 'x has to be a number!!!'
    if x > 0:
        print('x is positive')
    elif x == 0:
        print('x is 0')
    else:
        print('x is negative')
````

````python
sign_checker('test')
````

## 4. Loops

Loops are an absolutely essential part of any coder's toolkit. They can also be a bit hard to understand, so let's try to break it down.

### for loops

The most classic type of loop is the for loop. For our purposes, the for loop will do a task for every index in an iterable object (usually a list). Let's start by breaking down a super basic example.

````python
particles = ['proton', 'neutron', 'electron', 'neutrino']

for i in particles:
    print(i)
````

The first line is simple enough, I'm just creating a list with some particle names. The next line initiates the for loop. Essentially, the loop goes through the list of particles, and for every index in the list, I am printing it. Let's expand a bit on this example.

````python
particles = ['proton', 'neutron', 'electron', 'neutrino']

for i in particles:
    print(i)
    print('some random dummy step')
````

Notice, the for loop does everything within the indent chunk before moving on to the next index in the list. 

Let's do another example with some numbers. The goal is to create a for loop that can add up a sequence of numbers.

````python
numbers = [1,2,3,4,5]

x = 0
for n in numbers: 
    x = x + n 
    print(x)
    
````

Once again, the first line simply defines a list of numbers going from 1 to 5. Next, I define some variable x to be 0. We will sequentially add to this variable x for every index in the list of numbers. If it's helpful, think of x as like an empty bucket that will keep being added to for each step in the loop. Now in the for loop, notice that I redefine x in every step. Specifically, I redefine it to be the previous value of x plus the current index value that we're on. Finally, I print this value. You can do some quick mental math to see that the series worked properly.

### while loops

While loops are very similar to for loops. They perform an action while some condition is met. For example,

````python
i = 0
while i < 3:
    print(i)
    i = i + 1
````

In this case, we start from 0 and print integers until 3. This is specified in the condition i < 3. 

*VERY IMPORTANT NOTE*: Make sure you do not create a while loop where the condition is never met! It will indeed go on infinitely, and you're computer will freak out. Let's look at a horrible horrible example below (which I am too afraid to turn into real code).

```
i = 0
while i > -3:
    print(i)
    i = i + 1
```

In this case, we start at i = 0 and increase by 1 at each step of the loop. However, the loop goes on while i > -3. This will always be true! Hence, the while loop never stops...

### range() function

A really important built-in function that we use often is the range() function. It essentially just makes a list from 0 to the specified input *minus 1*. For example, if I do range(5), it will go from 0 to 4. You will use the range() function often. Do not forget that it only includes up to and not including the specified value!

````python
for i in range(5):
    print(i)
````

What is this good for? Why does it start at 0 and not include the final value? Here's a simple example: what if I just want to print happy birthday 5 times?

````python
for i in range(5):
    print('Happy Birthday')
````

Notice in this example, we never actually used i. We really just used it along with the range(5) function to note that we want the indented part to be performed 5 times.

Okay, let's do a more practical example. Keep track, because now I am using multiple lists!

````python
particle = ['proton', 'neutron', 'electron', 'neutrino']
charge = [1, 0, -1, 0]
mass = [938, 0.5, 940, 'ummmmm']

for i in range(4):
    print(particle[i], charge[i], mass[i])
````

In this example, notice that I had 3 lists of information: particle names, charges, and masses. Our goal was to print the name, charge, and mass of each respective particle on their own line. If we wanted to do this by hand, we would have needed 4 lines of print statements, each indexing the right value. Instead, we tactfully use the range() function to loop through the index values.

### break

The reserved word "break" will literally break the loop. This pairs nicely with conditionals. 

Let's print the particle names until we get to electron.

````python
particle = ['proton', 'neutron', 'electron', 'neutrino']
for i in particle:
    if i == 'electron':
        break
    else:
        print(i)
````

### continue

The reserved word "continue" is good for skipping over a step. Essentially, if continue is ever reached, it will skip the remainder of the indent chunk and go on to the next step in the for loop. 

Let's say this time, I want to skip 'electron'. Well, break is no good here because then it will stop at 'electron' but then also miss 'neutrino'. So, instead we use continue

````python
particle = ['proton', 'neutron', 'electron', 'neutrino']
for i in particle:
    if i == 'electron':
        continue
    print(i)
````

### Nested Loops (Advanced)

Like most things in Python, we can nest things by just adding more indents! Let's see a nested for loop in action.

````python
dead_stars = ['white_dwarf', 'neutron_star', 'black_hole']

for name in dead_stars:
    print(name)
    for n in range(5):
        print(n)
````

In this case, another for loop is being performed at each step of the original "outer" for loop. So, everytime a dead star is named, the inner for loop is counting from 0 to 4. 

Here's another example

````python
particle = ['proton', 'neutron', 'electron', 'neutrino']
infos = [[1, 938, 'hadron'], [0, 0.5, 'hadron'], [-1, 940, 'lepton'], [0, 'ummm', 'lepton']]
infos_header = ['Charge', 'Mass', 'Particle Type']

for i in range(4):
    print(particle[i])
    for n in range(3):
        print(infos_header[n], ':', infos[i][n])
    print()
````

This piece of code is a bit complicated, so look through it slowly on your own time until it makes sense. It comes for loops, the range() function, print statements, and indexing. Essentially, for each particle, we loop through each of their lists and prints the info.

## 5. Try-Except

Remember all the errors we have dealt with? For example, the classic divide by 0.

````python
5 / 0
````

Try-Except is like a special type of conditional, where we can account for errors.

````python
x = 5
y = 0

try:
    x / y
except:
    print('Cannot Divide by 0!')
````

We can also specify the type of error (just in case we want to deal with multiple possible errors). 

````python
x = 5
y = 0

try:
    x / y
except ZeroDivisionError:
    print('Cannot Divide by 0!')
````

This can be very good with debugging or error handling, which we might passively talk about later on. Just know for now that it exists and is similar to conditional formatting.
